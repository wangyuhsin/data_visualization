```python
import pandas as pd
```

```python
df = pd.read_csv('data/UTF8_Output.csv')
df.head()
```

```python
series_unique = df['Country'].value_counts()
series_unique
```

```python
df_unique = pd.DataFrame({'country':series_unique.index, 'count':series_unique.values})
df_unique = df_unique.head(20)
df_unique
```

```python
from bokeh.plotting import figure, ColumnDataSource
from bokeh.models.tools import HoverTool
from bokeh.transform import factor_cmap
from bokeh.palettes import Blues256, linear_palette
from bokeh.embed import components
from bokeh.io import output_notebook, show
```

```python
output_notebook()
```

```python
source = ColumnDataSource(df_unique)
```

```python
x = source.data['country']
x
```

```python
y = source.data['count']
y
```

```python
p = figure(x_range=x.tolist(), x_axis_label='country', plot_width=1000, plot_height=400,
          tools='pan,box_select,zoom_in,zoom_out,save,reset',
          toolbar_location='above',
          title='country & patent')
p.xgrid.grid_line_color = None
```

```python
p.vbar(source=source, x='country', top='count', width=0.4,
      fill_color=factor_cmap(
        "country",
        palette=linear_palette(Blues256, 105),
        factors=x
        ),
       fill_alpha=0.8,
       color=None
      )
```

```python
hover = HoverTool()
hover.tooltips = [('country', '@country'), ('patents count','@count')]
p.add_tools(hover)
show(p)
```

```python
script, div = components(p)
```
